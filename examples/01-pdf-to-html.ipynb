```python
from ushmm import pdf_to_images, images_to_text, clean_texts, remove_footers, process_testimony_texts

from datasets import load_dataset
from huggingface_hub import hf_hub_download, list_repo_files

from tqdm import tqdm

import os
```

```python
repo_id = "placingholocaust/ushmm-pdfs"
pdfs = list_repo_files(repo_id, repo_type="dataset")
len(pdfs)
```

```python


for pdf_filename in tqdm(pdfs[1:3]):
    # Remove 'pdfs/' from the filename if it exists
    clean_filename = pdf_filename.replace('.pdf', '')
    
    # Define the path for the output HTML file
    output_html_path = f"../data/{clean_filename}.html"
    
    # Check if the output HTML file already exists
    # if not os.path.exists(output_html_path):
    if "pdfs" in pdf_filename:
        hf_hub_download(repo_id="placingholocaust/ushmm-pdfs", filename=pdf_filename, local_dir="../data", repo_type="dataset")
        
        # Use the clean_filename for all subsequent operations
        images = pdf_to_images(f"../data/{pdf_filename}", f"../data/{clean_filename}/images", save=True)
        cropped_images = remove_footers(f'../data/{clean_filename}/images', output_directory=f'../data/{clean_filename}/images_cropped', save=True)
        texts = images_to_text(f"../data/{clean_filename}/images_cropped", save=True, output_folder=f"../data/{clean_filename}/text")
        texts = clean_texts(f"../data/{clean_filename}/text", save=True, output_directory=f"../data/{clean_filename}/clean_text")
        html_result = process_testimony_texts(f'../data/{clean_filename}/clean_text', output_file=output_html_path, save=True)
        print(f"Processed {clean_filename} and saved HTML output.")
    else:
        print(f"Skipping {clean_filename} as HTML output already exists.")
```
